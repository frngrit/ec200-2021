```python
new_date = []
for date in df['date']:
    try:
        new_date.append(datetime.strptime(date, "%Y-%m-%d %H:%M:%S"))
    except:
        new_date.append(np.nan)
```

```python
df['date'] = new_date
```

```python
def tokenize_words(txt):
    try:
        eiei = txt.lower().replace('\'','').split()
        new_text = [t for t in eiei if t.isalpha()]
        return new_text
    except:
        return None
```

```python
df['cleaned_text'] = [tokenize_words(t) for t in df['text']]
```
