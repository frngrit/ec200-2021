```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cleaned_data.csv',low_memory=False)
```

```python
df.drop(['is_retweet'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=['user_verified'],drop_first=True)
```

```python
df.head()
```

```python
df.set_index(['date'],inplace=True)
```

```python
clf_dict = {}
for date,tweet in zip(df.index,df['tweets']):
    try:
        clf_dict[date][0].extend(eval(tweet))
    except:
        clf_dict[date] = []
        clf_dict[date].append([])
        clf_dict[date][0].extend(eval(tweet))
```

```python
to_ana = pd.DataFrame()
to_ana['tweets'] = pd.DataFrame(clf_dict.values(),index=clf_dict.keys())
to_ana
```

```python
clf_dict = {}
for date,values in zip(df.index,df['user_followers']):
    try:
        clf_dict[date].append(values)
    except:
        clf_dict[date] = []
        clf_dict[date].append(values)
```

```python
mean_dict = {}
for date in clf_dict :
    mean_dict[date] = np.mean(clf_dict[date])
```

```python
to_ana['user_followers_mean'] = pd.DataFrame(mean_dict.values(),index=mean_dict.keys())
to_ana
```

```python
clf_dict = {}
for date,values in zip(df.index,df['user_favourites']):
    try:
        clf_dict[date].append(values)
    except:
        clf_dict[date] = []
        clf_dict[date].append(values)
```

```python
mean_dict = {}
for date in clf_dict :
    mean_dict[date] = np.mean(clf_dict[date])
```

```python
to_ana['user_favourites_mean'] = pd.DataFrame(mean_dict.values(),index=mean_dict.keys())
to_ana
```

```python
def myfunc(HR_return) :
    if HR_return < -0.2 :
        return 0
    elif HR_return < -0.1:
        return 1
    elif HR_return < 0.0 :
        return 2
    elif HR_return < 0.1 :
        return 3
    elif HR_return < 0.2 :
        return 4
    else:
        return 5
```

```python
vfunc = np.vectorize(myfunc)
```
